#### Importing necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

#### Importing the csv file

```python
lung_cancer=pd.read_csv("lung_cancer.csv")
lung_cancer
```

#### Checking for missing files

```python
lung_cancer.isna().sum()
```

#### Feature Encoding

```python
lung_cancer.replace({"LUNG_CANCER":{'YES':1,'NO':0},"GENDER":{"M":1,"F":0}},inplace=True)
lung_cancer
```

#### Normalization

```python
# apply normalization techniques
for column in lung_cancer.columns:
	lung_cancer[column] = (lung_cancer[column] - lung_cancer[column].min()) / (lung_cancer[column].max() - lung_cancer[column].min())	

# view normalized data
lung_cancer
```

#### Separating into Target(Y) and Non Target Variable(X)

```python
X=lung_cancer.drop('LUNG_CANCER',axis=1)
y=lung_cancer['LUNG_CANCER']
```

```python
from sklearn.model_selection import train_test_split

# Separating into train and testing
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

#### Feature Selection

```python
X_train.corr()
```

```python
import seaborn as sns
#Using Pearson Correlation
plt.figure(figsize=(12,10))
cor = X_train.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train, 0.7)
len(set(corr_features))
```

```python
corr_features
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

np.random.seed(42)

LRmodel= LogisticRegression()
LRmodel.fit(X_train, y_train)


print(f"Training Accuracy of Logistic Regression Model is {LRmodel.score(X_train,y_train)}")
print(f"Test Accuracy of Logistic Regression Model is {LRmodel.score(X_test,y_test)}")
```

#### DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier

np.random.seed(42)

dtcModel=DecisionTreeClassifier()

dtcModel.fit(X_train,y_train)

print(f"Training Accuracy of DecisionTreeClassifier Model is {dtcModel.score(X_train,y_train)}")
print(f"Test Accuracy of DecisionTreeClassifier Model is {dtcModel.score(X_test,y_test)}")
```

#### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

np.random.seed(42)

rfcModel=RandomForestClassifier()

rfcModel.fit(X_train,y_train)

print(f"Training Accuracy of RandomForestClassifier Model is {rfcModel.score(X_train,y_train)}")
print(f"Test Accuracy of RandomForestClassifier Model is {rfcModel.score(X_test,y_test)}")
```
